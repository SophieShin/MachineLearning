# 수업시간 내용
 - 1 hidden layer
 - 2 nodes
 - activation function : hidden layer에만 sigmoid
 - 이진 분류문제

```python
import math
import random
```

```python
# 초기값 랜덤화

a = random.uniform(-1, 1) # -1 ~ 1 까지 uniform distribution에서 random하게 한 값을 저장
b = random.uniform(-1, 1)
c = random.uniform(-1, 1)
d = random.uniform(-1, 1)
e = random.uniform(-1, 1)
f = random.uniform(-1, 1)
g = random.uniform(-1, 1)
h = random.uniform(-1, 1)
i = random.uniform(-1, 1)



eta= 0.0001 # 학습률
```

```python
print(a, b, c, d, e, f, g, h, i)
```

```python
def sigmoid(x):            # activation function인 sigmoid 정의
  return 1.0/(1.0 + math.exp(-x))
```

```python
# 5개의 데이터와 target 라벨지(binary-실패/성공)

data = [(5,10), (7,18), (10,8), (15,15), (20, 3), (23,12) ]
labels = [-1, -1, +1, +1, -1, -1]
```

```python
for counter in range(100000):
  da= db= dc= dd= de= df= dg= dh= di = 0.0 # gradient 초기화
  for idx in range(len(data)):
    x = float(data[idx][0])
    y = float(data[idx][1])
    z1 = a*x + b*y + c
    X = sigmoid(z1)
    z2 = d*x + e*y + f
    Y = sigmoid(z2)
    z = g*Y + h*Y + i
    Z = labels[idx]
    dz = 2*(z-Z)
    dz1 = dz*g*sigmoid(z1)*(1-sigmoid(z1))
    dz2 = dz*h*sigmoid(z2)*(1-sigmoid(z2))
    da+= dz1*x
    db+= dz1*y
    dc+= dz1
    dd+= dz2*x
    de+= dz2*y
    df+= dz2
    dg+= dz*X 
    dh+= dz*Y
    di+= dz

  a = a - eta*da
  b = b - eta*db
  c = c - eta*dc
  d = d - eta*dd
  e = e - eta*de
  f = f - eta*df
  g = g - eta*dg
  h = h - eta*dh
  i = i - eta*di


zValue = [0,0,0,0,0,0]
FinalValue=[0,0,0,0,0,0]

for idx in range(len(data)):
  x = float(data[idx][0])
  y = float(data[idx][1])
  z1 = a*x + b*y + c
  X = sigmoid(z1)
  z2 = d*x + e*y + f
  Y = sigmoid(z2)
  z = g*X + h*Y + i
  zValue[idx] += z

print('a=%1f, b=%1f, c=%1f, d=%1f, e=%1f, f=%1f, g=%1f, h=%1f, i =%1f' % (a, b, c, d, e, f, g, h, i))
print(zValue)
```
